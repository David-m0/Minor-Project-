# **Problem Statement** :





## Develop a program which analyzes the marks of a given set of students to assign grades and extract insights




---



# Logic :




### The following program uses Numpy and Python to analyze marks of a set of students based on the input given by the user.



> It accepts inputs from user to create a database of marks to analyze using input statements lists and numpy arrays

> It assigns a grade [A-F] based on the average marks obtain for each student using for loop and conditional statements

> It also calculates the highest and lowest marks in the given data set using Mathematical Operations






# Code :

```python

# Assumption: Let us assume that the maximum obtainable marks for each student is 500 in 5 subjects with 100 marks each

# Step 1: We first start by requesting the user to input the marks of students which will be used as the data base



import numpy as np

print("Welcome Lets Grade some Marks!\n")

def Marks_Grader():
  S1_M = int(input("Enter the total marks of student 1: "))
  S2_M = int(input("Enter the total marks of student 2: "))
  S3_M = int(input("Enter the total marks of student 3: "))
  S4_M = int(input("Enter the total marks of student 4: "))
  S5_M = int(input("Enter the total marks of student 5: "))

  Marks_ls = []
  Marks_ls.extend([S1_M,S2_M,S3_M,S4_M,S5_M])

  Mark_Array1 = np.array(Marks_ls)
  print(f"\nThe recorded marks of students are as follows: {Mark_Array1} \n")


Marks_Grader()


# Step 2: Now that we have data to work with we can calculate the average marks for each student

Average_ls = []

A_M1 = int(S1_M/5)
print(f"Avg Marks of Student 1: {A_M1}")
A_M2 = int(S2_M/5)
print(f"Avg Marks of Student 2: {A_M2}")
A_M3 = int(S3_M/5)
print(f"Avg Marks of Student 3: {A_M3}")
A_M4 = int(S4_M/5)
print(f"Avg Marks of Student 4: {A_M4}")
A_M5 = int(S5_M/5)
print(f"Avg Marks of Student 5: {A_M5}")

Average_ls.extend([A_M1,A_M2,A_M3,A_M4,A_M5])

print("\nThe average marks are stored and graded respectively: ", Average_ls)





# Step 3: Now we assign a grade based on the average marks of students

print("\n")
for avgm in Average_ls:
  if avgm >= 85:
    print("Grade A")
  elif avgm >= 75:
    print("Grade B")
  elif avgm >= 50:
    print("Grade C")
  elif avgm >= 20:
    print("Grade D")
  elif avgm < 20:
    print("You have failed.")

#Step 4: Displaying the highest and lowest among average marks

highest_avg = max(Average_ls)
lowest_avg = min(Average_ls)

print("\nThe Highest Average marks are: ", highest_avg)
print("\nThe Lowest Average marks are: ", lowest_avg)

# SAMPLE OUTPUT:
```
